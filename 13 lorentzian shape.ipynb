```python
import numpy as np
import matplotlib.pyplot as plt

# Define Lorentzian function
def lorentzian(x, x0=0, gamma=1):
    return (0.5 * gamma) / ((x - x0)**2 + (0.5 * gamma)**2) / np.pi

# Generate x values
x = np.linspace(-10, 10, 1000)

# Parameters
x0 = 0       # Peak center
gamma = 2    # Full width at half maximum (FWHM)

# Generate y values
y = lorentzian(x, x0, gamma)

# Plotting
fig, ax = plt.subplots()
ax.plot(x, y, label=f'Lorentzian ($x_0$={x0}, $\Gamma$={gamma})')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Lorentzian Curve')
ax.legend()
ax.grid(True)

# Save as SVG
plt.savefig("Figures/lorentzian.svg", format="svg")

print("✅ Lorentzian SVG saved as 'lorentzian.svg'")
```
